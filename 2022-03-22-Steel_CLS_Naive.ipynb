```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split,cross_validate

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

print(tf.__version__)

df = pd.read_excel('/content/Test_dataset.xlsx')
df.set_index('Date', inplace=True)
print(df.columns)

# df.info()
# split_time = 420

```

```python
Y_colname = ['w_inc_UD']
# X_remove = ['datetime', 'DateTime', 'temp_group', 'casual', 'registered']
# X_colname = [x for x in raw_fe.columns if x not in Y_colname+X_remove]
X_colname = ['Target', 'MA5','1)철광석-중국', '4)철스크랩-터키수입가', '5)WTI', '5)BSI-해운지수', '2)중국-Flat재고', '3)중국-판재+롱재고', '4)중국-석탄내수', 
             '1)열연-미국', '1)열연-뭄바이', '1)열연-북유럽', '1)열연-한국', '1)GI-상해', '1)GI-미국']



# 예측값 : 다음주 평균가격

# Dataset = df[X_colname+Y_colname]
Dataset = df[Y_colname]
Dataset = Dataset.dropna(axis=0)  # 데이터 없는거 지우기 (앞부분)



```

```python
ndf = Dataset[:]

ndf['Naive_shift'] = ndf['w_inc_UD'].shift(1)
# print(ndf)

ndf = ndf.loc[ndf.index >= '2019-12-30',:]  # 데이터 슬라이싱 (다른 것과 동일한 기준)
# Dataset = Dataset.dropna(axis=0)  # 데이터 없는거 지우기 (앞부분)
print(ndf)
```

```python
print("< 분류에 대한 Naive Forecasting : Shift > - 1주후 평균가 up/down\n")


score = ( np.mean(ndf['w_inc_UD'] == ndf['Naive_shift']) ) *100

print(" * 테스트 기간\n")
print(" - 정확도(%) :", '%0.1f' % score, "%")

```

```python
# 전체 기간에 대한 Naive forecasting (번외편)

ndf = Dataset[:]

ndf['Naive_shift'] = ndf['w_inc_UD'].shift(1)
# print(ndf)

# ndf = ndf.loc[ndf.index >= '2019-12-30',:]  # 데이터 슬라이싱 (다른 것과 동일한 기준)
# Dataset = Dataset.dropna(axis=0)  # 데이터 없는거 지우기 (앞부분)
# print(ndf)


Dataset['Naive_shift'] = Dataset['w_inc_UD'].shift(1)
Dataset = Dataset.dropna(axis=0)  # 데이터 없는거 지우기 (앞부분)

print("\n< 분류에 대한 Naive Forecasting : Shift > - 1주후 평균가 up/down\n")


score = ( np.mean(Dataset['w_inc_UD'] == Dataset['Naive_shift']) ) *100

print(" * 전체 기간")
print(" - 정확도(%) :", '%0.1f' % score, "%\n")
```
