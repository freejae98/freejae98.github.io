```python

# warning msg 무시
import warnings
warnings.filterwarnings('ignore')



############################################################################################################

# import win32com.client as win32
import pandas as pd 
import numpy as np
# import matplotlib.pyplot as plt    # 그래프를 그리는 패키지

# from matplotlib import font_manager, rc
# font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/Malgunsl.ttf").get_name()
# rc('font', family=font_name)

############################################################################################################


pd.options.display.float_format = '{:,.2f}'.format
pd.options.display.max_rows = 100
pd.options.display.max_columns = 50


# 기본 처리
# excel = win32.dynamic.Dispatch('Excel.application')
# excel.Visible = False #True
# excel.DisplayAlerts = False #True




# =============================================================================
# 
# 
# 1-1) 데이터 확인 - df 형태로 변환 : "1. Data_Lake.xlsx" 파일 - ndf1 (1번이 new data)
# 
# 
# =============================================================================

#파일 - 데이터프레임으로 받아오기
# fname1 = r"D:/★프로세스 혁신★/22.01월/1. AI 업체, 데이터분석 조사/4. 데이터 연결/MLOps - 운영관리/Daily Steel Monitoring_엑셀.xlsb"      
fname1 = r"/content/drive/MyDrive/1. Data_Lake.xlsx"


df1 = pd.read_excel(fname1)      
            
# df1 = df1.astype({'Date':'datetime64[ns]'})
# df1.set_index('Date', inplace=True)
# df = df.astype('float')



# wb1 = excel.Workbooks.Open(fname1, False, True, None)


# #데이터 프레임으로 저장, 각 시트 데이터를 df 변환
# ws1 = wb1.Worksheets(1)
# print(ws1.name)
# df1 = pd.DataFrame(ws1.UsedRange())
# df1.info()
# # print("basic_dataset.xlsx")
# print(df1)


# 엑셀은 닫기 (메모리 정리)
# wb1.Close(SaveChanges=False)  



# =============================================================================
# # Data frame 형태로 전환 - 데이터 정리 목적
# =============================================================================

# 1) ndf1 전처리 해주고.
df1.info()
df1.head(10)

# ndf1 = df1.iloc[5:,1:]
# ndf1 = df1.iloc[1:,:]


# ndf1.columns = df1.iloc[0,:]
# print(ndf1.columns)

ndf1 = df1.astype({'Date':'datetime64[ns]'})
ndf1.set_index('Date', inplace=True)

# ndf1.describe(include='all')
# ndf1 = ndf1.drop('6)Contract', axis=1)


ndf1 = ndf1.astype('float')

ndf1.info()




# =============================================================================
# 
# 
# 1-2) 데이터 확인 - df 형태로 변환 : "(STEP_1) 원본.xlsx" 파일 - ndf2 (2번은 예전 데이터)
# 
# 
# =============================================================================

#파일 - 데이터프레임으로 받아오기
# fname2 = r"D:/★프로세스 혁신★/22.09월/철강 MI 실행 추진/철강MI_데이터 수집/(STEP_1) 원본.xlsx"   # 처음에 1회만 해주는 것
fname2 = r"/content/drive/MyDrive/(Total) Data_Lake.xlsx"


df2 = pd.read_excel(fname2) 


# wb2 = excel.Workbooks.Open(fname2, False, True, None)


#데이터 프레임으로 저장, 각 시트 데이터를 df 변환
# ws2 = wb2.Worksheets(1)
# print(ws2.name)
# df2 = pd.DataFrame(ws2.UsedRange())
# df2.info()
# print("basic_dataset.xlsx")
# print(df2)


# 엑셀은 닫기 (메모리 정리)
# wb1.Close(SaveChanges=False)  



# =============================================================================
# # Data frame 형태로 전환 - 데이터 정리 목적
# =============================================================================

# 1) ndf1 전처리 해주고.
df2.info()
df2.head(10)

# ndf1 = df1.iloc[5:,1:]
# ndf2 = df2.iloc[1:,:]


# ndf2.columns = df2.iloc[0,:]
# print(ndf1.columns)

ndf2 = df2.astype({'Date':'datetime64[ns]'})
ndf2.set_index('Date', inplace=True)

# ndf1.describe(include='all')
# ndf1 = ndf1.drop('6)Contract', axis=1)


ndf2 = ndf2.astype('float')

ndf2.info()


# ndf2.columns

# =============================================================================
# 
# 
# 2-1) 데이터 합치기 : ndf2(기존) + ndf1(신규) = ndf3
#  - ndf1의 칼럼이름을 ndf2 기준으로 먼저 변경해줘야 하고 (태블로 연결용)
#  - ndf2에 없는 date에 대해서만 합칠 때 반영해줘야 함, 그리고 일부 계산서식은 재계산 해줘야 함 : MA5 등
#
# 
# =============================================================================

# ndf1(신규)의 칼럼 이름 변경


# 칼럼이름 참고파일 - 데이터프레임으로 받아오기-------------------------------------
fname3 = r"/content/drive/MyDrive/(참고) 각 칼럼이름 변경-태블로 연결용_T.xlsx"

t_df = pd.read_excel(fname3)      

# wb3 = excel.Workbooks.Open(fname3, False, True, None)


#데이터 프레임으로 저장, 각 시트 데이터를 df 변환 - nt_df 생성
# ws3 = wb3.Worksheets("한줄로 통합")
# print(ws3.name)
# t_df = pd.DataFrame(ws3.UsedRange())
# t_df.info()


# ndf1 = df1.iloc[5:,1:]
nt_df = t_df.iloc[:,1:]

# nt_df.columns = t_df.iloc[0,:]
# nt_df = t_df[:]

nt_df.info()

nt_df = nt_df.astype('object')


# 바꿀 값으로 입력
t_key = list(nt_df.columns)
tt_value = list(nt_df.iloc[0])
# len(t_value)

t_value = []    # 리스트안에 있는 것들을 string으로 변환 
for i in tt_value:
    temp = str(i)
    t_value.append(temp)
    
    

col_dic = {}    # 바뀌는 칼럼 이름을 딕셔너리 형태로 저장

# 대문자 변환            
t_key = [i.upper() for i in t_key]
# t_key[0] = t_key[0].upper()


for i in range(0,len(t_key)):
    col_dic[t_key[i]] = t_value[i].upper()  # 영문 대문자 전환, 지울 것은 이미 칼럽 값이 'NAN'으로 들어가 있음 

#ndf1, ndf2의 칼럼도 대문자 전환 - 나중에 합치기 위해

back_value = list(ndf2.columns)     # 나중에 다시 최종 df의 column 변경용 (이건 값)

ndf1.columns = [i.upper() for i in list(ndf1.columns)]
ndf2.columns = [i.upper() for i in list(ndf2.columns)]

back_key = list(ndf2.columns)  # 이건 key로 사용

back_dic = {}

for i in range(0,len(back_key)):
    back_dic[back_key[i]] = back_value[i]


# ndf1의 칼럼이름 변경----------------------------------------
len(ndf1.columns)  #82
len(col_dic.keys())  #82


# t_old = set(ndf1.columns)
# t_new = set(col_dic.keys())
# # t_sum = set(df3.columns)

# len(t_old)
# len(t_new)
# # len(t_sum)

# temp = t_old - t_new
# temp = t_new - t_old


for i in ndf1.columns:  #누락된게 있는지 확인용
    if i not in col_dic.keys():
        print(i)

new_col = []

for i in range(0,len(ndf1.columns)):
    temp = col_dic[ndf1.columns[i]]
    new_col.append(temp)

len(new_col)

#df 진화, 칼럼 이름 통일하여 교체하기
nndf1 = ndf1[:]
nndf2 = ndf2[:]
nndf1.columns = new_col



# ndf1 + ndf2 합쳐서 ndf3 생성----------------------------------------
# df3 = pd.concat([ndf2, ndf1])
# df3 = pd.merge(nndf2, nndf1, left_index=True, right_index=True, how='outer')

# nndf2 = nndf2.reset_index()
# nndf1 = nndf1.reset_index()

# concat 앞서서 중복 열 제거 (NAN 의미)
print(nndf1.columns[nndf1.columns.duplicated(keep=False)])
nndf1 = nndf1.drop('NAN', axis=1)

# old 데이터에 우선순위를 두고 통합 
df3 = pd.concat([nndf2, nndf1])   
df3 = df3.loc[~df3.index.duplicated(keep='first')]


ndf3 = df3.sort_index(ascending=True)



# =============================================================================
# 
# # 1) Date 재설정 + ffill 채우기 + MA5 칼럼 만들기 / 2) ML 예측/값 업데이트 하기
# 
# =============================================================================


# =============================================================================
# # Date 재설정
# =============================================================================
start_date = pd.to_datetime('2015-12-20') ## 시작 날짜  (MA5 만들기 위해 좀더 앞으로 땡김)

# end_date = pd.to_datetime(date.today() - timedelta(days=1)) ## 마지막 날짜 (어제 날짜로 기준 정함)  → 수정필요 : ndf1_2의 마지막 date 값으로.
end_date = ndf3.index[-1]


## 시작 날짜와 마지막 날짜 사이에 날짜를 생성하고 이를 인덱스로 설정
t_date = pd.date_range(start_date,end_date,freq='d')

t_date = pd.DataFrame(t_date, columns=['Date'])
t_date = t_date.astype({'Date':'datetime64[ns]'})

t_date.index.nunique()  #2346


for i in ndf3.columns:   #119개
    
    aa = ndf3[i].dropna(axis=0)
    bb = aa.loc[~aa.index.duplicated(keep='first')]
    cc = bb.sort_index(ascending=True)

    cc = cc.reset_index()       
    cc = cc.astype({'Date':'datetime64[ns]'})
    
    t_date = pd.merge(t_date, cc, how='left', on='Date')    # 이걸 쓰면 됨


t_date.set_index('Date', inplace=True)


sum_df1 = t_date[:]

# 중복 확인
sum_df1.index.nunique() == len(sum_df1)


# =============================================================================
# forward fill 해주고, 주말 일자는 빼주고.
# =============================================================================

sum_df2 = sum_df1.reset_index()    # 정수 인덱스 초기화


# 빈데이터 채우기
sum_df2 = sum_df2.fillna(method='ffill')
sum_df2.isnull().sum()

# 주말 데이터는 삭제 : mo_fill_wd
sum_df2['week'] = sum_df2['Date'].dt.dayofweek

sum_df3 = sum_df2[sum_df2['week'] < 5]

# 'week' 칼럼 삭제
sum_df3 = sum_df3.drop(['week'], axis=1)

sum_df3.set_index('Date', inplace=True)




# =============================================================================
# MA5 만들기, 빈데이터 채우기
# =============================================================================

# 일단 지저분한 칼럼들도 같이 가는걸로 하자. finance data 나중에 보완하게 될 것이고. 

sum_df4 = sum_df3[:]
# sum_df4 = sum_df3.drop(['Year', 'Quater', 'Quater_ver2', 'Month', 'Day'], axis=1)



# 칼럼 이름 바꾸기 : "_"를 " "로 일괄 변경
# temp = "아_름_다_운_강__산"
# temp = temp.replace("_", " ")
# temp

# for i in ds_sum_6.columns:
#     val = i.find("_")
#     if val != -1:
#         temp = i.replace("_", " ")
#         print(i + " - " + temp)
#         ds_sum_6 = ds_sum_6.rename(columns={i:temp})


# 예측대상 변수 연결 
predict_what = '2)열연-중국천진'


sum_df4['MA5'] = sum_df4[predict_what].rolling(5).mean()       # ★ 머신러닝 예측대상 설정 ★ - 변수 연결 : predict_what 
sum_df5 = sum_df4.dropna(subset=['MA5'], how='any', axis=0)


# #5일치 MV 구해서 집어넣기
# mo_fill_wd['MA5'] = mo_fill_wd['Target'].rolling(5).mean()


# RMB 환율이 '0'으로 들어간 row 삭제하기 + 특정 칼럼의 값이 '0'이면 이전 숫자로 채우기 (수식 오류에 대한 대처법) 
index1 = sum_df5[sum_df5['1)환율-RMB'] == 0].index

sum_df6 = sum_df5.drop(index1)

# sum_df5.columns[:116]       # 적용 대상이 되는 칼럼 리스트
len(sum_df6.columns)  #116

for i in sum_df6.columns[:116]:
    
    sum_df6.loc[sum_df6[i] == 0, i] = np.nan


# 빈데이터 채우기
sum_df6.isnull().sum()
sum_df7 = sum_df6.fillna(method='ffill')
sum_df7.isnull().sum()


# =============================================================================
# 최종 df의 칼럼 이름 원복함 - 태블로 연결용
# =============================================================================
new_col = []

for i in range(0,len(sum_df7.columns)):
    temp = back_dic[sum_df7.columns[i]]
    new_col.append(temp)

len(new_col)

#df 진화, 칼럼 이름 통일하여 교체하기
sum_df7.columns = new_col


# 파일 쓰기 - (total)Data_Lake
Total_file = '/content/drive/MyDrive/(Total) Data_Lake.xlsx'

sum_df7.to_excel(Total_file, index=True)

```
